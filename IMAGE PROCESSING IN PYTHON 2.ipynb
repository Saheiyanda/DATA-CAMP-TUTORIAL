In this exercise, you will compare both types of thresholding methods (global and local), to find the optimal way to obtain the binary image we need.

explain with given examples on how to create a mask for an image need to be restored in python application for image detection
To create a mask for an image that needs to be restored in a Python application for image detection, you can use various image processing techniques. One common approach is to use thresholding techniques to segment the image and create a binary mask.

Here is an example using the OpenCV library in Python to create a mask for an image using simple thresholding:

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('input_image.jpg', 0)  # Read image in grayscale

# Apply thresholding to create a mask
ret, mask = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Invert the mask
mask = cv2.bitwise_not(mask)

# Apply morphological operations to clean up the mask
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Apply the mask to the original image
restored_image = cv2.bitwise_and(image, image, mask=mask)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Restored Image', restored_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

In this example, we load an input image, apply simple thresholding to create a binary mask, invert the mask, and clean it up using morphological operations. Finally, we apply the mask to the original image to restore the desired image and display the results.

You can further enhance this process by using advanced image processing techniques or machine learning algorithms for more complex image restoration tasks.

````python
import cv2
import numpy as np

# Load the image
image = cv2.imread('image_with_logo.png', 0)  # Read image in grayscale

# Apply thresholding to create a mask
ret, mask = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Invert the mask
mask = cv2.bitwise_not(mask)

# Apply morphological operations to clean up the mask
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Apply the mask to the original image
restored_image = cv2.bitwise_and(image, image, mask=mask)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Restored Image', restored_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

# Perspective
In this exercise, you will detect the corners of a building using the Harris corner detector. The threshold_rel parameter will specify the minimum intensity of peaks.

````python
# Apply the detector  to measure the possible corners
# Import the corner detector related functions and module
from skimage.feature import corner_harris, corner_peaks
building_image = plt.imread('building_image.jpg')
# Convert image from RGB-3 to grayscale
building_image_gray = color.rgb2gray(building_image)

measure_image =  corner_harris(building_image_gray)

# Find the peaks of the corners using the Harris detector
coords = corner_peaks(measure_image, min_distance=200, threshold_rel=0.02)

# Show original and resulting image with corners detected
plt.imshow(building_image)
plt.title('original image')
````

````python
plt.imshow(measure_image)
plt.title('measure image')
````

````python
plt.imshow(coords)
plt.title('cord')
````

coords[:,1] refers to selecting all the values from the second column (index 1) of the coords array.
coords[:,0] refers to selecting all the values from the first column (index 0) of the coords array.
For example, if coords is a 2D array like this:

import numpy as np

coords = np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
coords[:,1] would extract [2, 4, 6], i.e., all the values from the second column.
coords[:,0] would extract [1, 3, 5], i.e., all the values from the first column.
In the context of plotting in 'plt.plot(coords[:,1],coords[:,0],'+r',markersize=15)', these values from specific columns are likely to be used as the x (horizontal) and y (vertical) coordinates for the plotted data points.

````python
def show_image_with_corners(image, coord,title='corners detected'):
    plt.imshow(image, interpolation='nearest',cmap='gray')
    plt.title(title)
    plt.plot(coords[:,1],coords[:,0],'+r',markersize=15)
    plt.axis('off')
    plt.show()
````
